## All NBA 2020 PREDICTION

### In this project, I'm going to use NBA per game data(2014-2020) from basketball reference to predict 2020 ALL NBA players.Here ALL NBA players refer to combination of ALL NBA first team+second team+thrid team. I gathered data from 2014-2019 ALL NBA teams and use per game statistics from 2014 to 2020 to predict ALL NBA 2020 teams

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

### 1.Importing per game data from basketballreference.com

```python
###selecting years between 2014 and 2020
years = np.arange(2014,2021,1)
years
```

```python
base_url = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
```

```python
seasons_per_game = pd.DataFrame()
```

```python
for year in years:
    req_url = base_url.format(year)
    req = requests.get(req_url)
    soup = BeautifulSoup(req.content, 'html.parser')
    table = soup.find('table', {'id':'per_game_stats'})
    df = pd.read_html(str(table))[0]
    df['Year'] = year
    
    seasons_per_game = seasons_per_game.append(df)
```

```python
##Take a look at our initial data
seasons_per_game
```

### 2.Data manipulation and cleaning

```python
###removing team,player name,pos and rank column
nbadata=seasons_per_game.iloc[:,3:31]
nbadata1=nbadata.reset_index()
nbadata2=nbadata1.iloc[:,0:28]
nbadata3=nbadata2.iloc[:,1:28]
nbadata4=nbadata3.drop(['Tm'],axis=1)
nbadata4=nbadata4.fillna(0.0)
```

```python
# A look of new data after some manipulations
nbadata4
```

```python
###Creating a new dataframe to mark ALL NBA players from 2014-2019
newd=seasons_per_game[['Player','Year']]
newd.to_csv('label.csv', sep='\t')
```

## 3 Model Building and prediction

### 3.1 KNN

```python
from sklearn import neighbors
```

```python
knn_predict = neighbors.KNeighborsClassifier(n_neighbors=1).fit(nbatrain_norm, allnba_train.ALLNBA).predict(nbatest_norm)
```

```python
sum(knn_predict)
```

### KNN Algorithm with neighbor=1 gives us 10 players for 2020 ALL NBA teams

```python
allnba_test.iloc[knn_predict == 1]
```

### 3.2 SVM

```python
from sklearn import svm
```

```python
svm_predict = svm.SVC().fit(nbatrain_norm, allnba_train.ALLNBA).predict(nbatest_norm)
```

```python
sum(svm_predict)
```

### SVM gives us 8 players for 2020 ALL NBA teams

```python
allnba_test.iloc[svm_predict == 1]
```

### 3.3 Naive Bayes

```python
from sklearn import naive_bayes
```

```python
nb_predict = naive_bayes.GaussianNB().fit(nbatrain_norm, allnba_train.ALLNBA).predict(nbatest_norm)
```

```python
sum(nb_predict)
```

### Naive Bayes gives us 59 players for 2020 All NBA teams

```python
allnba_test.iloc[nb_predict == 1]
```
